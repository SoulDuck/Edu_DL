# 컴퓨터가 이미지를 인식하는 방법

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open('cifar_horse.png')
print('Image size : {}'.format(img.size))
plt.imshow(img)
plt.show()


```

```python
# img 을 행렬로 바꾸어 줍니다.
np_img = np.asarray(img)
# Red Channel , Blue Channel , Green Channel 을 분리합니다. 
red_img = np_img[:,:,0]
blue_img = np_img[:,:,1]
green_img = np_img[:,:,2]
```

```python
# Extract Red Color
red_img = np_img.copy()
red_img[:,:,1]=0
red_img[:,:,2]=0
plt.imshow(red_img)
plt.show()
```

```python
green_img = np_img.copy()
green_img[:,:,0]=0
green_img[:,:,2]=0
plt.imshow(green_img)
plt.show()
```

```python
blue_img = np_img.copy()
blue_img[:,:,0]=0
blue_img[:,:,1]=0
plt.imshow(blue_img)
plt.show()
```
